# SVM classification

```python
import pandas as pd
import numpy as np
```

```python
from sklearn.datasets import load_breast_cancer
data = load_breast_cancer()
```

```python
data.data
```

```python
data.feature_names
```

```python
data.target
```

```python
data.target_names
```

```python
df = pd.DataFrame(np.c_[data.data, data.target], columns=np.append(data.feature_names, ['target']))
df.head(2)
```

```python
X = df.iloc[:, 0:-1]
y = df.iloc[:, -1]
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=51)
```

```python
from sklearn.svm import SVC
svm_rbf = SVC(kernel='rbf')
svm_rbf.fit(X_train, y_train)
```

```python
svm_rbf.score(X_test, y_test)
```

```python
svm_linear = SVC(kernel='linear')
svm_linear.fit(X_train, y_train)
```

```python
svm_linear.score(X_test, y_test)
```

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
sc.fit(X_train)
X_train_sc = sc.transform(X_train)
X_test_sc = sc.transform(X_test)
```

```python
svm_rbf2 = SVC(kernel='rbf')
svm_rbf2.fit(X_train_sc, y_train)
```

```python
svm_rbf2.score(X_test_sc, y_test)
```

```python
svm_linear2 = SVC(kernel='linear')
svm_linear2.fit(X_train_sc, y_train)
```

```python
svm_linear2.score(X_test_sc, y_test)
```

```python
X_test_sc[0]
```

```python
svm_linear2.predict([X_test_sc[0]])
```

```python
y_prid = svm_linear2.predict(X_test_sc)
```

```python
y_prid
```

```python
y_test
```

```python
from sklearn.metrics import mean_squared_error
import numpy as np
mse = mean_squared_error(y_test, y_prid)
rmse = np.sqrt(mse)
mse
rmse
```
